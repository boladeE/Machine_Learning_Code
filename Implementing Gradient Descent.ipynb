UNIVARIATE LINEAR REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('seaborn-data-master/seaborn-data-master/tips.csv')
```

```python
x_axis = data['total_bill'][0:50]
y_axis = data['tip'][0:50]
```

```python
plt.scatter(x_axis, y_axis)
plt.show()
```

```python
def mean_square_error(y_predicted, y_true):
    cost = np.sum((y_true-y_predicted)**2) / (2*len(y_true))
    return cost
```

```python
def derivative(x, y, w, b):
    m = len(x)
    
    for i in range(m):
        w_derivative = ((((w*x[i] + b) - y[i]) * (x[i])) / m)
        b_derivative= ((((w*x[i] + b) - y[i])) / m)
                  
    return w_derivative, b_derivative
```

```python
def gradient_descent(x, y, w, b):
    alpha = 0.05
    cost_at_each_iteration = []
    for i in range(1000):
        w_temp = w - (alpha*derivative(x, y, w, b)[0])
        b_temp = b - (alpha*derivative(x, y, w, b)[1])
        
        w = w_temp
        b = b_temp
        
        cost_at_each_iteration.append(mean_square_error((w*x + b), y))
    return w, b, cost_at_each_iteration
        
        
```

```python
m, c, cost = gradient_descent(x_axis, y_axis, 0.5, 0.5)
```

```python
y_predicted = [(i*m + c) for i in x_axis]
```

```python
plt.plot(x_axis, y_predicted, color='red')
plt.scatter(x_axis, y_axis)
```

```python
print(f'The intercept is {c}, and the gradient is {m}')
```

```python
print(f'The error is {mean_square_error(y_predicted, y_axis)}')
```

```python
no_of_iterations = [i for i in range(1000)]
plt.plot(no_of_iterations[0:20], cost[0:20])
plt.show()
```
